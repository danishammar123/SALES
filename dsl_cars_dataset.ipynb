**Q4:(Remove unwanted records**)
remove all records(rows) where weight is above 4000

```python
cars['Weight'] > 4000
```

```python
#we dont use above syntex becuase we dont need our result in boolean format
#so we use following syntex
cars[cars['Weight'] > 4000]
```

```python
#for removing above 4000 we use condition method ~
cars[~(cars['Weight'] > 4000)]
```

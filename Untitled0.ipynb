```python
# Feature Engineering
df.head()
```

```python
df['Surname'] = df['Name'].apply(lambda x:x.split()[0][0:-1])
df.head()
```

```python
df['Title'] = df['Name'].apply(lambda x:x.split()[1])
df.head()
```

```python
df.drop(columns=['Name'],inplace=True)
df.head()
```

```python
df['Has Cabin'] = df['Cabin'].apply(lambda x:'No' if x == 'No Cabin' else 'Yes')
df.head()
```

```python
df['family'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
def family_type(number):
    if number == 1:
        return 'alone'
    elif number > 1 and number <5:
        return 'small'
    else:
        return 'large'
```

```python
df['family_type'] = df['family'].apply(family_type)
```

```python
df.drop(columns=['SibSp','Parch','Cabin','family'],inplace=True)
df.head()
```

```python
df['family_type'].value_counts()
```
